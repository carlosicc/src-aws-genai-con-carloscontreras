# Prepara los Datos

```python
import sqlite3
conn = sqlite3.connect("demo.db")
c = conn.cursor()
```

```python
# Inserta filas
c.execute("CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT);")
c.execute("INSERT INTO users (name) VALUES ('Alice'), ('Bob');")
c.execute("INSERT INTO users (name) VALUES ('Carlos'), ('Pedro');")
```

```python
results = c.execute("select name from users;")
for r in results.fetchall():
    print(r)
```

```python
conn.commit()
conn.close()
```
