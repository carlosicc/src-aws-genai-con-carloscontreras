## Vamos a usar Guardrails con Strands Agents
> Prereq: Lanza notebook ```9.2. Guardrail API Lab 02```, hasta la creación de la primer versión

````python
#  # Copia y pega el ID de tu guardrail
notebook_guardrail_id = "72vl5apfxqxf" 
notebook_guardrail_version = "1" 
````

##### Nota: Miren cómo usamos el endpoint global en el model ID

````python
import json
from strands import Agent
from strands.models import BedrockModel

# Create a Bedrock model with guardrail configuration
bedrock_model = BedrockModel(
    model_id="global.anthropic.claude-sonnet-4-5-20250929-v1:0",
    guardrail_id=notebook_guardrail_id, 
    guardrail_version=notebook_guardrail_version,
    guardrail_trace="enabled",
)

# Create agent with the guardrail-protected model
agent = Agent(
    system_prompt="Eres un asistente amigable.",
    model=bedrock_model,
)

# Use the protected agent for conversations
response = agent("¿Cuánto tengo que invertir para mi retiro,si quiero generar $5,000 USD al mes?")

# Handle potential guardrail interventions
if response.stop_reason == "guardrail_intervened":
    print("Content was blocked by guardrails, conversation context overwritten!")

print(f"Conversation: {json.dumps(agent.messages, indent=4)}")
````

#### Recuerda borrar el Guardrail, si no lo vas a usar!

````python
import boto3
bedrock_client = boto3.client('bedrock')

# Borramos
response = bedrock_client.delete_guardrail(
    guardrailIdentifier=notebook_guardrail_id
)
print(f"Borrando guardrail creado en region:{boto3.Session().region_name}. Response status code: {response['ResponseMetadata']['HTTPStatusCode']}")
````
